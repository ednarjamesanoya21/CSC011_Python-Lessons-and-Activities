1. Define a function `sum_of_num` that takes a positive integer `n` and then returns the sum of all numbers from `1` to `n` using a `while` loop.

```python
def sum_of_num(n):
    i = 1
    sum = 0
    while i <= n:
        sum += i
        i += 1
    return sum

n = int(input("Enter number: "))
print(f"Sum of numbers from 1 to {n} : {sum_of_num(n)}")
```

2. Define a function `factorial` that takes an integer `n` and calculates and displays its factorial using a `while` loop. The factorial of a non-negative integer `n` is the product of all positive integers less than or equal to `n`.

```python
def factorial(n):
    result = 1

    if n == 0 or n == 1:
        print(f"Factorial of {n} : {result}")
    else:
        i = 2
        while i <= n:
            result *= i
            i += 1
        print(f"Factorial of {n} : {result}")

n = int(input("Enter number: "))
factorial(n)
```

3. Define a function `fibonacci` that would generate and display the Fibonacci series up to `n` terms. The Fibonacci sequence starts with `0` and `1`, and each subsequent number is the sum of the previous two.

```python
def fibonacci(n):
    fib0 = 0
    fib1 = 1
    i = 0
    result = 0

    while i <= n:
        if i == 0:
            result = fib0
        elif i == 1:
            result = fib1
        else:
            fibn = fib0 + fib1
            result = fibn
            fib0 = fib1
            fib1 = fibn
        print(result, end=" ")
        i += 1

n = int(input("Enter how many terms: "))
fibonacci(n)
```

4. Define a function `sum_of_digits` that returns the sum of digits of a given number. For example, if the input is `1234`, the output should be `10` because `1+2+3+4 = 10`.

```python
def sum_of_digits(n):
    sum = 0

    while n > 0:
        digit = n % 10
        sum += digit
        n = n // 10

    return sum

n = int(input("Enter number: "))
print(f"Sum of digits of number {n} : {sum_of_digits(n)}")
```

5. Define a function `draw_triangle` that takes in `h` as the height of the triangle and displays a patterns of stars (*) in the shape of a right-angled triangle.

```python
def draw_triangle(h):
    row = 1
    while row <= h:
        column = 0
        while column < row:
            print("*", end="")
            column += 1
        print("")
        row += 1

h = int(input("Enter height of triangle: "))
draw_triangle(h)
```

6. Define a function `show_multiplication_table` that displays the multiplication table (up to `10`) of a given number `n`.

```python
def show_multiplication_table(n):
    i = 1
    while i <= 10:
        result = i * n
        print(f"{n} x {i} = {result}")
        i += 1

def show_multiplication_table_v2(n):
    for i in range(1, n+1):
        for j in range(1, 11):
            result = i * j
            print("{0:^4}".format(result), end="")
        print("")


n = int(input("Enter number: "))
show_multiplication_table_v2(n)
```

7. Define a function `except_multiples_of_five` that displays all numbers from `1` to `n` except multiples of `5` using a `while` loop and `continue` statement.

```python
def except_multiples_of_five(n):
    i = 1
    while i <= n:
        if i % 5 == 0:
            i += 1
            continue
        else:
            print(i, end=" ")
            i += 1

n = int(input("Enter number: "))
except_multiples_of_five(n)
```

8. Define a function `fizz_buzz`that takes in an integer `n` and prints all numbers from `1` to `n`. For multiples of `3`, print `"Fizz"` instead of the number. For multiples of `5`, print `"Buzz"`. For numbers which are multiples of both `3` and `5`, print `"FizzBuzz"`.

```python
def fizz_buzz(n):
    i = 1
    while i <= n:
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz")
        elif i % 3 == 0:
            print("Fizz")
        elif i % 5 == 0:
            print("Buzz")
        else:
            print(i)
        i += 1

n = int(input("Enter number: "))
fizz_buzz(n)
```

9. Define a function `sum_of_input` that takes in `n` as the number of times the user will be asked to input a number, and then displays the sum of the numbers provided by the user.

```python
def sum_of_input(n):
    sum = 0
    i = 0
    while i < n:
        number = int(input("Enter number: "))
        sum += number
        i += 1
    print(sum)

n = int(input("How many times should we ask the user for an input? "))
sum_of_input(n)
```

10. Define a function `is_prime` that takes in a positive integer `n` and iterates through divisors from `2` to the square root of `n`. If `n` is less than `2`, the function returns `False`, because prime numbers are greater than or equal to `2`. If any divisor evenly divides `n`, the function returns `False`, it means that `n` is not prime. If no divisors are found, the function returns `True`, indicating that `n` is prime.

```python
def is_prime(n):
    if n < 2:
        return False

    div = 2
    while div <= (n **.5):
        if n % div == 0:
            return False
        div += 1

    return True

n = int(input("Enter number: "))
is_prime(n)
```
