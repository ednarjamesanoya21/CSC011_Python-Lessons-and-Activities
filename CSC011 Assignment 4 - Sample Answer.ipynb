1. Define a function `count_vowels` that takes in a string and returns a count of vowels `(a, e, i, o, u)` in the string. Ignore case sensitivity.

```python
def count_vowels(str):
    vowels = "aeiouAEIOU"
    count = 0
    for c in str:
        if c in vowels:
            count += 1
    return count

user_input = input("Enter string: ")
count_vowels(user_input)
```

2. Define a function `reverse_str` that takes a string and displays the string reversed. For example, if the input is `"hello"`, the output should be `"olleh"`.

```python
def reverse_str(str):
    last_char_pos = len(str) - 1
    while last_char_pos >= 0:
        print(str[last_char_pos], end="")
        last_char_pos -= 1

user_input = input("Enter string: ")
reverse_str(user_input)
```

3. Define a function `replace_char` that takes a string, a character to find, and a character to replace it with, and displays the modified string with all occurrences of the first character replaced by the second character.

```python
def replace_char(str, target, replace):
    new_str = ""
    for c in str:
        if c == target:
            new_str += replace
        else:
            new_str += c
    print(new_str)

user_input = input("Enter string: ")
replace_char(user_input, "l", "w")
```

4. Define a function `count_char_frequency` that takes a string and prints the frequency of each character in the string.

```python
def count_char_frequency(str):
    unique_chars = ""
    for c in str:
        if c not in unique_chars:
            unique_chars += c

    for uc in unique_chars:
        count = 0
        for c in str:
            if uc == c:
                count += 1
        print(f"{uc} : {count}")

user_input = input("Enter string: ")
count_char_frequency(user_input)
```

5. Define a function `is_palindrome` that takes a string and returns `True` if the string is a palindrome (reads the same forwards and backwards), and `False` otherwise.

```python
def is_palindrome(str):
    len_str = len(str)
    i = 0
    li = len_str - 1

    while i < len_str:
        if str[i] != str[li]:
            return False
        i += 1
        li -= 1

    return True

user_input = input("Enter string: ")
is_palindrome(user_input)
```
