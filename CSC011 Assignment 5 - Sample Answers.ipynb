1. Define a function `sum_positive` that takes a list of integers and returns the sum of all positive numbers in the list.

```python
def sum_positive(numbers):
    sum = 0
    for number in numbers:
        if number % 2 == 0:
            sum += number
    return sum

some_numbers = [1,2,3,4,5]
print(f"Sum of positive numbers: {sum_positive(some_numbers)}")
```

2. Define a function `get_length` that takes a tuple of strings and returns a new tuple with the lengths of each string in the original tuple.

```python
def get_length(strings):
    lens = ()
    for string in strings:
        lens += (len(string),)
    return lens
some_tuples = ("a","whole","new","world")
print(get_length(some_tuples))
```

3. Write a function `get_highest_score` that takes a list of tuples, where each tuple contains a student's name and their corresponding score, and returns the name of the student with the highest score.

```python
def get_highest_score(student_score_list):
    highest_score = 0
    name_with_highest_score = ""
    for (name, score) in student_score_list:
       if score > highest_score:
            name_with_highest_score = name
    return name_with_highest_score

students_scores = [("John", 5), ("Paul", 10), ("George", 20), ("Ringo", 18)]
print(f"student with highest score is {get_highest_score(students_scores)}")
```

4. Define a function `min_max` that takes a list of integers and returns a tuple containing the indices of the minimum and maximum elements.

```python
import random

def generate_int_list(size, max_int):
    n_list = []
    for i in range(0, size):
        n_list.append(random.randint(0, max_int))
    return n_list

def min_max(numbers):
    min_index = 0
    max_index = 0
    min = numbers[min_index]
    max = numbers[max_index]


    for i in range(len(numbers)):
        if numbers[i] <= min:
            min = numbers[i]
            min_index = i

        if numbers[i] >= max:
            max = numbers[i]
            max_index = i

    return (min_index, max_index)

numbers = generate_int_list(7, 30)

print(numbers)
print(min_max(numbers))
```

5. Define a function `extract_first_char` that takes a tuple of strings and returns a list of the first characters of each string.

```python
def extract_first_char(tuple_str):
    first_chars = []
    for ts in tuple_str:
        first_chars.append(ts[0])
    return first_chars

print(extract_first_char(("John", "Paul","Ringo","George")))
```

Define a function `distance_from_origin` that takes a list of tuples, where each tuple contains two integers representing coordinates, and returns a new list of new tuples containing the tuple of coordinates and the computed distance from the origin `(0, 0)`. Define a helper function `calculate_distance` to calculate the distance of a single point from the origin using the Euclidean distance formula.

```python
def calculate_distance(point1, point2):
    point1_x, point1_y = point1 #unpack the tuple
    point2_x, point2_y = point2 #unpack the tuple
    return (((point2_x - point1_x) ** 2) + ((point2_y - point1_y) ** 2)) ** 0.5

def distance_from_origin(points):
    origin = (0,0)
    result = []
    for point in points:
        distance = calculate_distance(point, origin)
        result.append((point, distance))
    return result

print(distance_from_origin([(3, 0), (0, 4), (1, 1)]))
```
