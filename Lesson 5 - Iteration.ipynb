Computers are often used to automate repetitive tasks. Repeating identical or similar tasks without making errors is something that computers do well and people do poorly.

Repeated execution of a set of statements is called **iteration**. Because iteration is so common, Python provides several language features to make it easier.

Before we look at those features, let's do a quick review of the following ideas..

# Assignment

As we have mentioned previously, it is legal to make more than one assignment to the same variable. A new assignment makes an existing variable refer to a new value (and stop referring to the old value).

```python
airtime_remaining = 15
print(airtime_remaining)
airtime_remaining = 7
print(airtime_remaining)
```

It is especially important to distinguish between an assignment statement and a Boolean expression that tests for equality. Because Python uses the equal token (`=`) for assignment, it is tempting to interpret a statement like `a = b` as a Boolean test. Unlike mathematics, it is not! Remember that the Python token for the equality operator is `==`.

Note too that an **equality test is symmetric, but assignment is not**. For example, if `a == 7` then `7 == a`. But in Python, the statement `a = 7` is legal and `7 = a` is not.

# Updating variables

When an assignment statement is executed, the right-hand side expression (i.e. the expression that comes after the assignment token) is evaluated first. This produces a value. Then the assignment is made, so that the variable on the left-hand side now refers to the new value.

One of the most common forms of assignment is an update, where the new value of the variable depends on its old value.

```python
n = 5
n = 3 * n + 1
print(n)

```

```python
airtime_remaining = airtime_remaining - 4
```

Before you can update a variable, you have to **initialize** it to some starting value, usually with a simple assignment:

```python
score_player_1 = 0
score_player_1 = score_player_1 + 1
```

Updating a variable by adding 1 to it is called an **increment** of a variable,

```python
score_player_1 = score_player_1 - 1
```

Updating a variable by subtracting 1 is called a **decrement** of a variable.

# `while` Loop

A loop is a language construct that allows us to perform iteration or to repeatedly execute a statement or a group of statements until a terminating condition is satisfied

A while statement implements a while loop to perform iteration

```python
user_input_number = int(input("Some integer>"))
number = 2
if number <= user_input_number:
    print("hello ", number)

while number <= user_input_number:
    print(number)
    number = number + 1

print("This is a statement after the while loop")
```

The code above means, while `number` is less than or equal to the value of `user_input_number`, the body of the `while` loop will continue to execute. Within the body, the value of `number` is incremented, every iteration.

More formally,


*   Declare and initialize the loop control variables at `line 1 and line 2`
*   Evaluate the condition at `line 4`, yelding a value which is either `True` or `False`
*   If the value is `True`, execute each of the statements in the body (`line 5 and line 6`) and then go back to the `while` statement at `line 4`
*   If the value is `False`, exit the `while` statement and continue execution at the next statement which is `line 8` in this case



The body of the loop should change the value of one or more variables so that eventually the condition becomes false and the loop terminates. Otherwise the loop will repeat forever, which is called an **infinite loop**.

When using a while loop one has to manage the **loop control variable** yourself: give it an initial value, test for completion, and then make sure you change something in the body so that the loop terminates.

## Example 1: Counting digits

```python
def num_digits(n):
    count = 0
    while n != 0:
        count = count + 1
        n = n // 10
    return count

print(num_digits(12345))
```

```python
12345 // 10
```

```python
1234 // 10
```

```python
123 // 10
```

```python
12 // 10
```

```python
1 // 10
```

This function demonstrates an important pattern of computation called a **counter**. The variable `count` is initialized to `0` and then incremented each time the loop body is executed. When the loop exits, `count` contains the result — the total number of times the loop body was executed, which is the same as the number of digits.

## Example 2: Letting the user choose when to stop the loop

```python
prompt = "\nTell me something, and I will repeat it back to you. \nType 'quit' to end the program. "
message = ""

while message != "quit":
    message = input(prompt)
    print(message)
```

## Example 3: Using a flag

In the previous example, we had the program perform certain tasks while
a given condition was true. But what about more complicated programs in
which many different events could cause the program to stop running?

For a program that should run only as long as many conditions are true,
you can define one variable that determines whether or not the entire program is active. This variable, called a *flag*, acts as a signal to the program.

We can write our programs so they run while the flag is set to `True` and stop running when any of several events sets the value of the flag to `False`.

```python
prompt = "\nTell me something, and I will repeat it back to you. \nType 'quit' to end the program. "
active = True

while active:
    message = input(prompt)
    if message == 'quit':
        active = False
    else:
        print(message)
```

# `break` statement

The `break` statement is used to immediately leave the body of its loop. The next statement to be executed is the first one after the body:

```python
prompt = "\nTell me something, and I will repeat it back to you. \nType 'quit' to end the program. "

while True:
    message = input(prompt)

    if message == 'quit':
        break
    else:
        print(message)
```

# `continue` statement

This is a control flow statement that causes the program to immediately skip the processing of the rest of the body of the loop, for the current iteration. But the loop still carries on running for its remaining iterations:

```python
def display_even_numbers(n):
    number = -1 #2
    while number <= n: # 1 <= 5
        number += 1
        if number % 2 != 0: # 2 % 2 != 0
            continue
        print(number)


display_even_numbers(5)
```

# Loop behaviors

## pre-test loop (*standard behavior*)

When loops are designed that they do their tests at the start, before executing any part of the body. They’re called pre-test loops, because the test happens before (pre) the body.

break and return are our tools for adapting this standard behaviour.

## middle-test loop

When the the exit test of a loop is in the middle of the body, rather than at the beginning or at the end.

```python
total = 0
while True:
    response = input("Enter the next number. (Leave blank to end) ")
    if response == "":
        break
    total += int(response)
print("The total of the numbers you entered is ", total)
```

## post-test loop

The exit test is the last thing in the body

Post-test loops are used when you want to be sure that the loop body always executes at least once (because the first test only happens at the end of the execution of the first loop body)


```python
import random

def guessing_game():

    rng = random.Random()   #random number generator
    random_number = rng.randint(1, 5) # Get random number between 1 and 30.

    guesses = 0
    max_guess = 3
    msg = ""
    guessed = False

    while True:
        guess = int(input(msg + "\nGuess my number between 1 and 5: "))
        guesses += 1

        if guesses == max_guess:
            print("Game Over!!")
            break

        if guess > random_number:
            msg += str(guess) + " is too high.\n"
        elif guess < random_number:
            msg += str(guess) + " is too low.\n"
        else:
            guessed = True
            break
    if guessed:
        print("\n\nGreat your guess is {1}, you got it in {0} guesses!\n\n".format(guesses, guesses))
        print(f"\n\nGreat your guess is {guess}, you got it in {guesses} guesses!\n\n")

while True:
    guessing_game()
    response = input("Play again? (yes or no)")
    if response != "yes":
        break

print("Goodbye!")
```
