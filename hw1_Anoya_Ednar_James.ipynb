0. Define a function `draw_square()` that displays a 3x3 set of asterisks.

```python
def draw_square():
    for i in range(3):
        for j in range(3):
            print("*", end=" ")
        print()

# Call the function to draw the square
draw_square()
```

2. Write a program that prompts for two numbers and computes for the
sum, difference, product, and quotient of those numbers using fruitful functions defined for each operation. Display the results as shown in the example output below:

Screen Shot 2022-10-05 at 12.28.42 PM.png

```python
def get_numbers():
    num1 = float(input("What is the first number? "))
    num2 = float(input("What is the second number? "))
    return num1, num2

def calculate_sum(num1, num2):
    return num1 + num2

def calculate_difference(num1, num2):
    return num1 - num2

def calculate_product(num1, num2):
    return num1 * num2

def calculate_quotient(num1, num2):
    if num2 != 0:
        return num1 / num2
    else:
        return "Cannot divide by zero"

def main():
    num1, num2 = get_numbers()

    sum_result = calculate_sum(num1, num2)
    difference_result = calculate_difference(num1, num2)
    product_result = calculate_product(num1, num2)
    quotient_result = calculate_quotient(num1, num2)

    print(f"{num1} + {num2} = {sum_result}")
    print(f"{num1} - {num2} = {difference_result}")
    print(f"{num1} * {num2} = {product_result}")
    print(f"{num1} / {num2} = {quotient_result}")

if __name__ == "__main__":
    main()
```

3. Write a program to evenly divide pizzas. Define a void function `get_pizza_slices` that takes in the number of people, the number of pizzas, and the number of slices per pizza. Ensure that the number of pieces comes out even. Display the number of pieces of pizza each person should get. If there are leftovers, show the number of leftover pieces. Display the results as shown in the example output below:

Screen Shot 2022-10-05 at 12.25.28 PM.png

```python
def get_pizza_slices(people, pizzas, slices_per_pizza):
    total_slices = pizzas * slices_per_pizza
    slices_per_person = total_slices // people
    leftover_slices = total_slices % people

    print(f"{people} people with {pizzas} pizzas")
    print(f"Each person gets {slices_per_person} pieces of pizza.")

    if leftover_slices == 0:
        print("There are 0 leftover pieces.")
    else:
        print(f"There are {leftover_slices} leftover pieces.")


num_people = int(input("How many people? "))
num_pizzas = int(input("How many pizzas do you have? "))
slices_per_pizza = int(input("How many slices per pizza? "))


get_pizza_slices(num_people, num_pizzas, slices_per_pizza)
```

4. Write a program that asks the user to enter the radius of a circle and then computes and displays the circle’s area using a fruitful function `area_of_circle(radius)`. Use the formula

   `area = PI x radius x radius`

   where `PI` is the constant macro `3.14159`

```python
def area_of_circle(radius, PI):
    area = PI * radius**2
    return area

PI = 3.14159

num_radius = float(input("Give me the radius: "))
result = area_of_circle(num_radius, PI)
print("The area of the circle is:", result)
```

5. Write a program
that will ask for a
number of days
and then will
show how many
hours, minutes
and seconds are
in that number of
days. Define your own set of functions.

```python
def convert_days_to_hours_minutes_seconds(days):
    hours = days * 24
    minutes = hours * 60
    seconds = minutes * 60
    return hours, minutes, seconds

def main():
    try:
        num_days = float(input("Enter the number of days: "))
        if num_days < 0:
            raise ValueError("Number of days should be a non-negative value.")

        hours, minutes, seconds = convert_days_to_hours_minutes_seconds(num_days)

        print(f"{num_days} days is equal to:")
        print(f"{hours} hours")
        print(f"{minutes} minutes")
        print(f"{seconds} seconds")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()

```

6. Write a program that implements a fruitful function `get_gallons_of_paint` to determine the gallons of paint needed to paint the ceiling of a room. In your program, prompt the user for the length and width of the ceiling a room. Assume one gallon covers 400 square feet. Display the number of gallons needed to paint the ceiling as a whole number. Remember that you cannot buy a partial gallon of paint. You must round up to the next whole gallon. Display the results as shown in the example output below:

Screen Shot 2022-10-05 at 12.27.42 PM.png

```python
import math

def get_gallons_of_paint(length, width):
    square_feet = length * width
    gallons_needed = math.ceil(square_feet / 400)
    return gallons_needed

def main():
    try:
        length = float(input("Enter the length of the ceiling in feet: "))
        width = float(input("Enter the width of the ceiling in feet: "))

        if length <= 0 or width <= 0:
            raise ValueError("Length and width must be positive values.")

        gallons_needed = get_gallons_of_paint(length, width)

        print(f"You will need to purchase {gallons_needed} gallon(s) of paint "
              f"to cover {length * width} square feet.")

    except ValueError as e:
        print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
