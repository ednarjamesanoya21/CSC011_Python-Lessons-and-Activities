1. Write a program that prompts the user to enter an integer and implement a function `is_odd_even` and then determines whether the given number is even or odd.

```python
def is_odd_even(x):
    if x % 2 == 0:
        return "Even"
    else:
        return "Odd"

x = int(input("Input a number to determine if it is odd or even: "))
result = is_odd_even(x)
print(f"The number {x} is {result}.")
```

2. Write a program that prompts the user to enter the lengths of three sides of a triangle. Implement a function `check_triangle` that determines whether the triangle is equilateral, isosceles, scalene, or not a triangle at all.

```python
def check_triangle(side1, side2, side3):


    if side1 + side2 <= side3 or side1 + side3 <= side2 or side2 + side3 <= side1:
        return "Not a triangle"

    elif side1 == side2 and side2 == side3:
        return "Equilateral triangle"

    elif side1 == side2 or side1 == side3 or side2 == side3:
        return "Isosceles triangle"

    else:
        return "Scalene triangle"

side1 = float(input("1st Side: "))
side2 = float(input("2nd Side: "))
side3 = float(input("3rd Side: "))

print(check_triangle(side1, side2, side3))
```

3. Write a program to simulate a traffic light system. Prompt the user to enter the current state of the traffic light (red, yellow, or green). Implement a function `next_state` that determines the next state of the traffic late based on the input.


```python
def next_state(current_state):
    if current_state == "green":
        return "yellow"
    elif current_state == "yellow":
        return "red"
    elif current_state == "red":
        return "green"
    else:
        return "Invalid State"

current_state = input("Enter the current traffic light color to determine what's next (green, yellow, red): ").lower()

print(next_state(current_state), "is the next traffic light color.")
```
